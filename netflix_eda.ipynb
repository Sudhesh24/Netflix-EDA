Exploratory Data Analysis of Netflix Movies & TV Shows

Objective
The objective of this project is to analyze the Netflix Movies and TV Shows dataset using Python to understand content distribution, genre trends, release patterns, and country-wise production insights

```python
import pandas as pd
```

```python
df = pd.read_csv("netflix_titles.csv")
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.countplot(x='type', data=df)
plt.title("Movies vs TV Shows on Netflix")
plt.show()
```

```python
df['listed_in'].head()
```

```python
df['listed_in'].str.split(', ').explode().value_counts().head(10)
```

```python
top_genres = df['listed_in'].str.split(', ').explode().value_counts().head(10)
```

```python
top_genres.plot(kind='bar')
plt.title("Top 10 Genres on Netflix")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.show()
```

```python
df['release_year'].value_counts().sort_index().head()
```

```python
df['release_year'].value_counts().sort_index().plot()
plt.title("Netflix Content Growth Over Years")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.show()
```

```python
df['country'].head()
```

```python
top_countries = df['country'].dropna().str.split(', ').explode().value_counts().head(10)
top_countries
```

```python
top_countries.plot(kind='bar')
plt.title("Top 10 Content Producing Countries on Netflix")
plt.xlabel("Country")
plt.ylabel("Number of Titles")
plt.show()
```

In this project, exploratory data analysis was performed on the Netflix Movies and TV Shows dataset to understand content distribution and trends. The analysis showed that Netflix has a larger number of movies compared to TV shows. Drama and International Movies are the most dominant genres on the platform. Content production has increased significantly after 2016, indicating Netflixs rapid global expansion. The United States is the top content-producing country, followed by India and the United Kingdom.
